## 3. Make Pandas DataFrame and visualise it using Matplotlib and Seaborn

### Let us create Pandas DataFrame from the Iris dataset and perform various operation on it to explore, analyze, and understand the dataset's content.

```python
# Create a Pandas DataFrame (df) from the "data" key of the loaded Iris dataset. 
# It assigns the feature names from the "feature_names" key as column names for the DataFrame. 
# Essentially, it's converting the Iris dataset's data into a tabular format with labeled columns.

df = pd.DataFrame(data["data"], columns = data["feature_names"])
```

```python
# list the column names in the DataFrame.

df.columns
```

```python
# Provide a summary of the DataFrame's information, including the number of non-null entries in each column and
# the data types of the columns.

df.info()
```

```python
# Display the first 5 rows.
# It gives a preview of the structure and content of the DataFrame.

df.head()
```

```python
# Generate summary statistics for each numeric column in the DataFrame. 
# It includes metrics like mean, standard deviation, minimum, maximum, and quartiles. 
# This helps in understanding the distribution of the numerical features.

df.describe()
```

### Now, let us visualize the four different attributes (sepal length, sepal width, petal length, and petal width) of the Iris dataset using Matplotlib. The following code snippet creates a 2x2 grid of subplots, each displaying a line plot of one of the four attributes of the Iris dataset. Each subplot has a title that corresponds to the plotted attribute.

```python
# Draw 2x2 subplot and show the line graph for each of four attributes.

fig, ax =plt.subplots(2,2)
fig.tight_layout(pad=2)

ax[0, 0].plot(df["sepal length (cm)"])
ax[0, 0].set_title("sepal length (cm)")

ax[0, 1].plot(df["sepal width (cm)"])
ax[0, 1].set_title("sepal width (cm)")

ax[1, 0].plot(df["petal length (cm)"])
ax[1, 0].set_title("petal length (cm)")

ax[1, 1].plot(df["petal width (cm)"])
ax[1, 1].set_title("petal width (cm)")
```

### Histograms

```python
# Draw a 2x2 subplot and show the histogram of each of the four attributes.

fig, axes = plt.subplots(2, 2, figsize=(10,10))
 
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df['sepal length (cm)'], bins=7)
 
axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df['sepal width (cm)'], bins=5);
 
axes[1,0].set_title("Petal Length")
axes[1,0].hist(df['petal length (cm)'], bins=6);
 
axes[1,1].set_title("Petal Width")
axes[1,1].hist(df['petal width (cm)'], bins=6);
```

#### From the above plot, we can see that  

1. The highest frequency of the sepal length is between 30 and 35 which is between 5.5 and 6
2. The highest frequency of the sepal Width is around 70 which is between 3.0 and 3.5
3. The highest frequency of the petal length is around 50 which is between 1 and 2
4. The highest frequency of the petal width is between 40 and 50 which is between 0.0 and 0.5

### Next, we add a new column named "target" to the Pandas DataFrame (df) and populate it with the target values from the Iris dataset. Each target value corresponds to the species of the iris flower for the respective row in the DataFrame. By adding this column, the DataFrame now associates each sample with its target class.

```python
# Add "target" column in DataFrame to associate each sample with its target class.

df["target"] = data["target"]
```

```python
# Calculate and display the count of each unique value in the "target" column.
# Also, display the datatype of "target" column.

# Note:
# 0 -> setosa
# 1 -> versicolor
# 2 -> virginica

df["target"].value_counts()
```

### We can use Seaborn library to create a count plot that visualizes the distribution of target classes (species of iris flowers) in the DataFrame df derived from the Iris dataset. 

```python
# Make the count plot of DataFrame and show it.

sns.countplot(x='target', data=df, palette='bright',)
plt.show()
```

### Relation between variables

```python
# Compare Sepal Length and Sepal Width attributes.

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)',
                hue='target', data=df, palette='bright')
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```

#### From the above plot, we can infer that  

1. Species Setosa (shown by 0 label and blue color) has smaller sepal lengths but larger sepal widths.
2. Versicolor Species (shown by 1 label and orange color) lies in the middle of the other two species in terms of sepal length and width
3. Species Virginica (shown by 2 label and green color) has larger sepal lengths but smaller sepal widths.

```python
# Compare Petal Length and Petal Width attributes.

sns.scatterplot(x='petal length (cm)', y='petal width (cm)',
                hue='target', data=df, palette='bright')
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```

#### From the above plot, we can infer that  

1. Species Setosa (shown by 0 label and blue color) has smaller petal lengths and widths.
2. Versicolor Species (shown by 1 label and orange color) lies in the middle of the other two species in terms of petal length and width
3. Species Virginica (shown by 2 label and green color) has the largest of petal lengths and widths.

```python
# Plot all the columns relationships using a pairplot. 
# It can be used for multivariate analysis.

sns.pairplot(df, hue='target', height=2, palette='bright')
```

#### From the above plot, we can infer that  

Species Setosa has the smallest of petals widths and lengths. It also has the smallest sepal length but larger sepal widths. Such information can be gathered about any other species.

### Checking Numerical Column


```python
# Find the numerical columns.

numerical_columns = []
for column_name, data_type in zip(df.columns, data_types):
  if data_type != 'object':
    numerical_columns.append(column_name)
    
print('Numerical columns:', numerical_columns)
```

### Handling Correlation

```python
# Find the correlation using Pearson Correlation method.

# Calculates the Pearson correlation coefficient matrix for the columns in the DataFrame df, 
# showing the pairwise correlation between numeric attributes.
df.corr(method='pearson')
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr(method='pearson')

# Get the correlation values with respect to the target column
target_correlations = correlation_matrix['target'].drop('target')

# Get the top 4 correlations (positive and negative) with respect to the target
top_4_target_corr = target_correlations.nlargest(4)

# Print the top 4 correlations with respect to the target along with their corresponding columns
for column, correlation_value in top_4_target_corr.items():
    print(f"Correlation between '{column}' and 'target': {correlation_value:.4f}")
```

#### We can see that the petal length and petal width features have the highest correlation with the class label. It is 0.949 and 0.956 respectively. This means that these two features are the most important features for predicting the class label.
