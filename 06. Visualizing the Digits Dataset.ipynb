# 1. Import required libraries

```python
# Import the Scikit-learn datasets module and Pandas, Matplotlib, and Seaborn libraries.

from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the Dataset, check features, and targets.

```python
# Import the Digits dataset from the scikit-learn library.

data = datasets.load_digits()
```

```python
# Print the names of the features in the Digits dataset.

print(data.feature_names)
```

```python
# Print the target values of the Digits dataset.

print(data.target_names)
```

```python
# Print the shape of the target array.

print(data.target.shape)
```

```python
# Display digit 0.

# Sets the color map of Matplotlib to grayscale.
plt.gray()

# Displays a matrix representation of the first image in the dataset using grayscale colors.
plt.matshow(data.images[0]);
```

## 3. Make Pandas DataFrame and visualise it using Matplotlib and Seaborn

### Let us create Pandas DataFrame from the Digits dataset and perform various operation on it to explore, analyze, and understand the dataset's content.

```python
# Create a Pandas DataFrame (df) from the "data" key of the loaded Digits dataset. 
# It assigns the feature names from the "feature_names" key as column names for the DataFrame. 
# Essentially, it's converting the Digits dataset's data into a tabular format with labeled columns.

df = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
# list the column names in the DataFrame.

df.columns
```

```python
# Provide a summary of the DataFrame's information, including the number of non-null entries in each column and
# the data types of the columns.

df.info()
```

```python
# Display the first 5 rows.
# It gives a preview of the structure and content of the DataFrame.

# Set the display option to show all rows and columns without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df.head()
```

```python
# Generate summary statistics for each numeric column in the DataFrame. 
# It includes metrics like mean, standard deviation, minimum, maximum, and quartiles. 
# This helps in understanding the distribution of the numerical features.

df.describe()
```

### Next, we add a new column named "target" to the Pandas DataFrame (df) and populate it with the target values from the Digits dataset. Each target value corresponds to the digit for the respective row in the DataFrame. By adding this column, the DataFrame now associates each sample with its target class.

```python
# Add "target" column in DataFrame to associate each sample with its target class.

df["target"] = data["target"]
```

```python
# Calculate and display the count of each unique value in the "target" column.
# Also, display the datatype of "target" column.

df["target"].value_counts()
```

### We can use Seaborn library to create a count plot that visualizes the distribution of target classes (digits) in the DataFrame df derived from the Digits dataset. 

```python
# Make the count plot of DataFrame and show it.

sns.countplot(x='target', data=df, palette='bright',)
plt.show()
```

### Examine the DataFrame considering both features and target aspects


```python
# Display the first 5 rows (sixty-four features and target class).

df.head()
```

```python
# Display the last 5 rows (sixty-four features and target class).

df.tail()
```

```python
# Check the data type of each column.

data_types = df.dtypes

print(data_types)
```

### Checking Categorical Columns


```python
# Find the categorical columns.

categorical_columns = []
for column_name, data_type in zip(df.columns, data_types):
  if data_type == 'object':
    categorical_columns.append(column_name)
    
print('Categorical columns:', categorical_columns)
```

### Checking Numerical Column


```python
# Find the numerical columns.

numerical_columns = []
for column_name, data_type in zip(df.columns, data_types):
  if data_type != 'object':
    numerical_columns.append(column_name)
    
print('Numerical columns:', numerical_columns)
```

### Checking Missing Values


```python
# Find the missing values.

df.isnull().sum()
```

#### We can see that no column has any missing value.

### Handling Correlation

```python
# Find the correlation using Pearson Correlation method.

# Set the display option to show all rows and columns without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Calculates the Pearson correlation coefficient matrix for the columns in the DataFrame df, 
# showing the pairwise correlation between numeric attributes.
df.corr(method='pearson')
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr(method='pearson')

# Get the correlation values with respect to the target column
target_correlations = correlation_matrix['target'].drop('target')

# Get the top 10 correlations (positive and negative) with respect to the target
top_10_target_corr = target_correlations.nlargest(10)

# Print the top 10 correlations with respect to the target along with their corresponding columns
for column, correlation_value in top_10_target_corr.items():
    print(f"Correlation between '{column}' and 'target': {correlation_value:.4f}")
```

#### We can see the top ten pixel features that have the highest correlation with the class label. These features are the most important features for predicting the class label.
