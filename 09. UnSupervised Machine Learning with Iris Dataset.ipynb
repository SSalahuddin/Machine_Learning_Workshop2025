# This notebook performs Unsupervised Machine Learning on the Iris Dataset. It consists of five sections.

## 1. Import required libraries

```python
# Import libraries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score
```

## 2. Load the Dataset, check features, and targets.

```python
# Load the Iris dataset.
iris = load_iris()

# Features (sepal length, sepal width, petal length, petal width).
X = iris.data

# True labels (for evaluation, not used in training).
y = iris.target
```

## 3. Apply UnSupervised Machine Learning Algorithm


```python
# Apply K-means clustering.
# We assume 3 clusters (like the 3 species).
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Train the model.
kmeans.fit(X)
```

```python
# Get cluster labels and centroids.
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

## 4. Evaluate Performance


```python
# Evaluate clustering performance (optional, since it's unsupervised).
# Compare clusters with true labels (for demonstration).
print("Adjusted Rand Score (closer to 1 is better):", adjusted_rand_score(y, cluster_labels))
print("Silhouette Score (closer to 1 is better):", silhouette_score(X, cluster_labels))
```

## 5. Visualize results


```python
# Visualize clusters (using first two features for simplicity).
plt.figure(figsize=(12, 5))

# Plot clusters predicted by K-means.
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering (Predicted)')

# Plot true species labels for comparison.
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('True Species Labels')

plt.tight_layout()
plt.show()
```
